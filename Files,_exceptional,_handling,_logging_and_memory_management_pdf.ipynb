**11.What is memory management in Python**?

‚óè Ans:- Memory management in Python refers to the process of allocating and releasing
memory to ensure efficient use of resources while the program runs. Python handles
memory management automatically through its garbage collection mechanism, but it
also provides developers with tools to manage memory more effectively.

1. Automatic Memory Management (Garbage Collection)

‚óè Python uses automatic garbage collection to free up memory that is no longer in use
(i.e., when objects are no longer referenced).

‚óè The garbage collector identifies and removes objects that are not reachable,
preventing memory leaks.

2. Memory Allocation

‚óè When a variable or object is created, memory is allocated for it.

‚óè Python manages this allocation using the heap (for dynamically created objects) and
the stack (for function calls and local variables).

3. Reference Counting

‚óè Each object in Python has a reference count that tracks the number of references to
the object.

‚óè When the reference count drops to zero (i.e., no more references), the object is
automatically deallocated.

4. Garbage Collection (GC)

‚óè Python uses cyclic garbage collection to handle circular references (objects
referencing each other).

The gc module provides developers the ability to interact with the garbage collector if
needed.

**15. How does Python's garbage collection system work**?

‚óè Ans- üîπ How Python‚Äôs Garbage Collection System Works

‚óè Python‚Äôs garbage collection (GC) system automatically reclaims memory by removing
objects that are no longer needed, helping your program run efficiently without
memory leaks.

‚úÖ 1. Reference Counting (Primary Mechanism)

‚óè Every object in Python has a reference count ‚Äî the number of variables that point to
it.

‚óè When an object‚Äôs reference count drops to zero, Python automatically deletes it.

üß† Example:

a = [1, 2, 3] # Reference count = 1

b = a # Reference count = 2

del a # Reference count = 1

del b # Reference count = 0 ‚Üí Collected


‚úÖ 2. Garbage Collector for Cyclic References

‚óè Sometimes, objects reference each other in a cycle, so their reference count never
reaches 0 ‚Äî like:

a = {}

b = {}

a["b"] = b

b["a"] = a


‚úÖ 3. Generational Garbage Collection


Python divides objects into 3 generations:

‚óè Gen 0: New objects

‚óè Gen 1: Survived one GC cycle

‚óè Gen 2: Long-lived objects

‚úÖ 4. Manual Control (Optional)

‚óè You can interact with the GC using the gc module:

import gc

gc.collect() # Manually trigger garbage collection

gc.disable() # Turn off automatic GC (not recommended)

**23.What are the challenges associated with memory management in Python**?

‚óè Ans:- üîπ Challenges Associated with Memory Management in Python Although
Python automates memory management using its garbage collection and reference
counting systems, there are still several challenges developers might face.

‚ö†Ô∏è 1. Memory Leaks

‚óè Caused when objects are unintentionally kept alive due to lingering references.

‚óè Often occurs with circular references or global/static variables.

‚ö†Ô∏è 2. Inefficient Memory Usage

‚óè Python's high-level abstractions (lists, dictionaries, objects) may use more memory
than low-level languages like C.

‚óè Data structures can become memory-heavy if not optimized.

‚ö†Ô∏è 3. Circular References

‚óè Reference counting alone can‚Äôt handle them.

‚óè Python's garbage collector detects and collects these, but it's not always perfect or
fast.

‚ö†Ô∏è 4. Unreleased Resources

‚óè Objects like file handles, DB connections, or sockets consume memory if not
explicitly closed.

‚ö†Ô∏è 5. Global Variables

‚óè Keeping large objects in the global scope may delay garbage collection or increase
memory footprint unnecessarily.

‚ö†Ô∏è 6. Misuse of C Extensions or Third-party Libraries

‚óè Some C-based libraries may not integrate well with Python‚Äôs garbage collector,
leading to leaks.

‚ö†Ô∏è 7. Long-Lived Objects in Loops

‚óè Creating large objects repeatedly without clearing them or letting them expire can
cause memory bloat over time

```python
#6 Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

# Configure logging
logging.basicConfig(filename='error_log.txt', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError as e:
        logging.error("Attempted to divide by zero: %s", e)
        print("Error: Division by zero is not allowed.")

# Example usage
result = divide(10, 0)
```

```python
#13 How would you check if a file exists before attempting to read it in Python?
import os

file_path = "my_data.txt"

if os.path.exists(file_path):
    try:
        with open(file_path, "r") as file:
            content = file.read()
            print("File content:")
            print(content)
    except Exception as e:
        print(f"An error occurred while reading: {e}")
else:
    print(f"Error: The file '{file_path}' does not exist.")
```

```python
#14 Write a program that uses the logging module to log both informational and error messages?
import logging

# Configure the logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    filename='app.log',  # Logs will be saved to this file
    filemode='w'          # 'w' to overwrite the file each run; use 'a' to append
)

def divide(a, b):
    logging.info(f"Attempting to divide {a} by {b}")
    try:
        result = a / b
        logging.info(f"Result: {result}")
        return result
    except ZeroDivisionError as e:
        logging.error("Division by zero error")
        logging.exception("Exception occurred")  # Logs the stack trace

# Example usage
divide(10, 2)
divide(5, 0)
```

```python
#15 Write a Python program that prints the content of a file and handles the case when the file is empty?
def print_file_contents(filename):
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            if not contents:
                print("The file is empty.")
            else:
                print("File Contents:\n")
                print(contents)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
file_name = 'example.txt'
print_file_contents(file_name)
```

```python
#18 How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

# Create logger
logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)  # You can change this to INFO, WARNING, etc.

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",        # Log file name
    maxBytes=1 * 1024 * 1024,  # 1MB size limit
    backupCount=5     # Keep up to 5 backup files
)

# Create a formatter and set it for the handler
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example log messages
logger.debug("This is a debug message")
logger.info("This is an info message")
logger.warning("This is a warning message")
```

22 How can you check if a file is empty before
attempting to read its contents?

1. Using os.path.getsize():

You can check the size of the file. If the size is 0, the file is empty.

import os

file_path = "your_file.txt"

if os.path.getsize(file_path) > 0:

with open(file_path, "r") as file:

content = file.read()

print(content)

else:

print("The file is empty.")


2. Using os.stat():

Another way to get the size of the file is using os.stat(), which also provides information
about the file size.

import os

file_path = "your_file.txt"

if os.stat(file_path).st_size > 0:

with open(file_path, "r") as file:

content = file.read()

print(content)

else:

print("The file is empty.")


3. Checking if read() returns empty:

You can open the file and attempt to read its content directly. If read()
returns an empty
string, it means the file is empty.

with open("your_file.txt", "r") as file:

content = file.read()

if content:

print(content)

else:

print("The file is empty.")
